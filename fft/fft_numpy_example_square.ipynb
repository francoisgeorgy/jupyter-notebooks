```python
import numpy as np
import matplotlib.pyplot as plt

f_s = 10000.0 # Hz
f = 1.0 # Hz
time = np.arange(0.0, 3.0, 1/f_s/100)
x = 5 * np.sign(np.sin(2 * np.pi * f * time))

plt.plot(time, x)
plt.xlabel("Time (sec)")
plt.ylabel("x")
plt.show()
```

```python
fft_x = np.fft.fft(x)
n = len(fft_x)
freq = np.fft.fftfreq(n, 1/f_s)

#fft_x_shifted = np.fft.fftshift(fft_x)
#freq_shifted = np.fft.fftshift(freq)

half_n = int(np.ceil(n/2.0))
fft_x_half = (2.0 / n) * fft_x[:half_n]
freq_half = freq[:half_n]

plt.plot(freq_half, np.abs(fft_x_half))
plt.xscale('log')
plt.axis((0.001, 1, 0, 8))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.show()
```
