Playing with the number of samples and the sample spacing

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fft

samples = 1024           # Number of samplepoints; must be greater than 2 * max freq
spacing = 1.0 / 1024   # sample spacing

fig, (signal, fft) = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

x = np.linspace(0.0, samples * spacing, samples, endpoint=False)   # Return evenly spaced numbers over a specified interval.
y = np.sin(20.0 * 2.0*np.pi*x) + 0.5*np.sin(50.0 * 2.0*np.pi*x)
_ = signal.plot(x, y)
yf = scipy.fft.fft(y)
xf = scipy.fft.fftfreq(samples, spacing)[:samples//2]
_ = fft.semilogy(xf, 2.0/samples * np.abs(yf[0:samples//2]))
fft.set_xlim(left=0, right=100)
```
