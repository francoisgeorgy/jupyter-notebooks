```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fft


samples = 600            # Number of samplepoints
f = 10
# periods = 10.0
# spacing = (1.0/f * periods) / samples     # sample spacing
spacing = 1.0/180     # sample spacing

x = np.linspace(0.0, samples * spacing, samples, endpoint=False)

y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
# y = np.sign(np.sin(f * 2.0*np.pi*x))

# fig, (wave, wavefft) = plt.subplots(2, 1, figsize=(10,10))
# wave.plot(x, y)

yf = scipy.fft.fft(y)

# method 1:
# xf = np.linspace(0.0, 1.0/(2.0*spacing), samples//2)
# wavefft.plot(xf, 2.0/samples * np.abs(yf[:samples//2]))

# method 2:
xf = scipy.fft.fftfreq(samples, spacing)[:samples//2]
# wavefft.plot(xf, 2.0/samples * np.abs(yf[0:samples//2]))
plt.semilogy(xf, 2.0/samples * np.abs(yf[0:samples//2]))
# wavefft.axis((1, 100, 0, 1.2))

# print(2.0/samples * np.abs(yf[:samples//2]))
# print(ff)

# plt.plot(xf, 2.0/N * np.abs(yf[:N//2]))
# plt.axis((20, 200, 0, 1.2))
# plt.xscale('log')
plt.grid()
plt.show()
```
