```python
import matplotlib.pyplot as plt
import numpy as np
import math
#plt.rcParams["figure.figsize"] = (8,8)


A = 1.0
T = 1.0
periods = 3
t = np.arange(0.0, periods*T, 0.001)     # Return evenly spaced values within a given interval.

sin = np.sin(2 * np.pi * t)
tri = (4*A/T) * abs( ((t-T/4) % T) - (T/2)) - A
square = (4*np.floor(t/T)) - (2*np.floor(2*(t/T))) + 1

fig, ax = plt.subplots(3, 1, figsize=(10,10))
ax[0].plot(t, sin)
ax[1].plot(t, tri)
ax[2].plot(t, square)

ax[0].set(xlabel='time', ylabel='amplitude', title='Sine')
ax[1].set(xlabel='time', ylabel='amplitude', title='Triangle')
ax[2].set(xlabel='time', ylabel='amplitude', title='Square')
ax[0].grid()
ax[1].grid()
ax[2].grid()

#fig.savefig("test.png")
fig.tight_layout()
plt.show()
```
