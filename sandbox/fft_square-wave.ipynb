```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack

N = 204800   # sample points
f = 20     # Hz
p = 2      # show 2 periods

t = np.linspace(0, p/f, N)    # Return evenly spaced numbers over a specified interval
y = np.sign(np.sin(2.0*np.pi*f*t))

fig, (wave, wavefft) = plt.subplots(2, 1, figsize=(10,10))
wave.plot(t, y)
wave.grid()

# wave_fft.plot(t, square)
# wave_fft.grid()

# fig.tight_layout()
# plt.show()

# sample spacing:
ss = p/f / N

# # sample spacing
# T = 1.0 / 800.0
# x = np.linspace(0.0, N*T, N)
# y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)

print(len(y))

yfft = scipy.fftpack.fft(y)
xfft = np.linspace(0.0, 1.0/(2.0* ss), N//2)

print(len(yfft))
print(len(xfft))

wavefft.plot(xfft, 2.0/N * np.abs(yfft[:N//2]))

# # fig, ax = plt.subplots()
# # ax.plot(xf, 2.0/N * np.abs(yf[:N//2]))
# plt.plot(xf, 2.0/N * np.abs(yf[:N//2]))
# plt.axis((20, 200, 0, 1.2))
# wavefft.axis((20, 200, 0, 1.2))
plt.xscale('log')
plt.grid()
plt.show()
```
