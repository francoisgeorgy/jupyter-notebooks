```python
import matplotlib.pyplot as plt
import numpy as np
import math
#plt.rcParams["figure.figsize"] = (8,8)

resolution = 500  # x-axis finesse; without unit; how many points to plot per period

f = 25.0      # frequency, Hz
A = 5.0       # half-amplitude, without unit; total amplitude will be from -A to +A
T = 1.0 / f   # period, s
periods = 4   # how many periods to show

t = np.arange(0.0, periods * T, T / resolution)     # Return evenly spaced values within a given interval.
#t = np.arange(0.0, periods / f, 1 / f / resolution)     # Return evenly spaced values within a given interval.

sin = A * np.sin(2 * np.pi * t * f)
tri = A * abs(((t-T/4) % T) - (T/2)) * 4 / T - A
square = A * np.sign(sin)

fig, ax = plt.subplots(3, 1, figsize=(10,10))
ax[0].plot(t, sin)
ax[1].plot(t, tri)
#ax[2].plot(t, square)
ax[2].plot(t, square)

#ax[0].set_xticklabels(np.arange(0.0, periods / f + 0.1, 0.1))
#ax[1].set_xticks(np.arange(0.0, periods / f + 0.1, 0.1))
#ax[2].set_xticks(np.arange(0.0, periods / f + 0.1, 0.1))
#ax[3].set_xticks(np.arange(0.0, periods / f + 0.1, 0.1))

ax[0].set(xlabel='time', ylabel='amplitude', title='Sine')
ax[1].set(xlabel='time', ylabel='amplitude', title='Triangle')
ax[2].set(xlabel='time', ylabel='amplitude', title='Square')
#ax[3].set(xlabel='time', ylabel='amplitude', title='Square2')
ax[0].grid()
ax[1].grid()
ax[2].grid()
#ax[3].grid()

# after plotting the data, format the labels
#current_values = ax[0].gca().get_xticks()
# using format string '{:.0f}' here but you can choose others
#plt.gca().set_yticklabels(['{:,.0%}'.format(x) for x in current_values])

#fig.savefig("test.png")
fig.tight_layout()
plt.show()
```
