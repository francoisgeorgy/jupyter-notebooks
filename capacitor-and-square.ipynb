```python
import matplotlib.pyplot as plt
import numpy as np
import math
#plt.rcParams["figure.figsize"] = (8,8)

micro = 1 / 1000000.0
kilo = 1000.0

A = 1.0
T = 1.0
periods = 3
t = np.arange(0.0, periods*T, 0.001)     # Return evenly spaced values within a given interval.

R = 10.0 * kilo
C = 10.0 * micro
#T = R * C
V0 = 1.0

sin = np.sin(2 * np.pi * t)
tri = (4*A/T) * abs( ((t-T/4) % T) - (T/2)) - A

square = (4*np.floor(t/T)) - (2*np.floor(2*(t/T))) + 1
#square = np.where(t%T, 1, 2)

#discharge = (4*np.floor(t/T)) - (2*np.floor(2*(t/T))) + 1

steps = np.floor(2*(t/T))
#discharge = np.floor(2*(t/T))
discharge = V0 * np.exp(- (t-steps/2) /(R*C))

#discharge = V0 * np.exp(-t/(R*C))
#discharge = np.where(square > 0, V0 * np.exp(-t/(R*C)), V0 * np.exp(-(t % T)/(R*C)))

#def func(x):
#    import numpy as np
#    x = np.where(x<0, 0., x*10)
#    return x   

fig, ax = plt.subplots(6, 1, figsize=(10,15))
ax[0].plot(t, discharge)
ax[1].plot(t, square)
ax[2].plot(t, square * discharge)
ax[3].plot(t, tri)
ax[4].plot(t, tri * discharge)
ax[5].plot(t, steps)

ax[0].set(xlabel='time', ylabel='amplitude', title='Discharge')
ax[1].set(xlabel='time', ylabel='amplitude', title='Square')
ax[2].set(xlabel='time', ylabel='amplitude', title='Square')
ax[3].set(xlabel='time', ylabel='amplitude', title='Triangle')
ax[4].set(xlabel='time', ylabel='amplitude', title='Triangle')
ax[5].set(xlabel='time', ylabel='amplitude', title='Steps')
ax[0].grid()
ax[1].grid()
ax[2].grid()
ax[3].grid()
ax[4].grid()
ax[5].grid()

#fig.savefig("test.png")
fig.tight_layout()
plt.show()
```
